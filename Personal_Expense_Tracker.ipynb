# Personal Expense Tracker

## Introduction

Managing personal finances effectively requires tracking income and expenses over time. This project provides a simple **Python-based Personal Expense Tracker** that helps users monitor monthly spending and calculate remaining budget.

The project demonstrates fundamental Python programming concepts such as:

- Variables and data types
- Lists and data structures
- Functions and modular code
- Basic financial calculations
- Formatted console output

It is designed as a beginner-friendly project for learning Python while solving a real-world problem.

```python
"""
Personal Expense Tracker
Author: Iliya Ishaya
Description: A simple Python program for tracking and summarizing monthly expenses.
"""

from typing import List


def calculate_total_expenses(prices: List[float]) -> float:
    """
    Calculate total expenses.

    Parameters
    ----------
    prices : list
        List of expense amounts.

    Returns
    -------
    float
        Total expenses.
    """
    return sum(prices)


def calculate_balance(budget: float, total_expenses: float) -> float:
    """
    Calculate remaining balance.

    Parameters
    ----------
    budget : float
        Monthly budget.
    total_expenses : float
        Sum of all expenses.

    Returns
    -------
    float
        Remaining balance.
    """
    return budget - total_expenses


def display_summary(user: str,
                    month: str,
                    currency: str,
                    budget: float,
                    total_expenses: float,
                    balance: float) -> None:
    """
    Display formatted expense summary.
    """

    print("=" * 40)
    print("        PERSONAL EXPENSE TRACKER")
    print("=" * 40)

    print(f"Owner: {user}")
    print(f"Period: {month}")

    print("-" * 40)

    print(f"Currency: {currency}")
    print(f"Starting Budget: {currency}{budget:.2f}")
    print(f"Total Expenses: {currency}{total_expenses:.2f}")
    print(f"Remaining Balance: {currency}{balance:.2f}")

    print("=" * 40)


def main():
    """Main program execution."""

    user = "John Doe"
    current_month = "February 2026"
    currency = "$"
    budget = 1000.00

    items = [
        "Rent",
        "Electricity",
        "Clothes",
        "Data Bundle",
        "KFC",
        "Netflix"
    ]

    prices = [
        400.00,
        80.00,
        100.00,
        50.00,
        90.00,
        25.00
    ]

    categories = [
        "Housing",
        "Utility",
        "Basic",
        "Utility",
        "Food",
        "Entertainment"
    ]

    necessary = [
        True,
        True,
        True,
        True,
        False,
        False
    ]

    total_expenses = calculate_total_expenses(prices)
    balance = calculate_balance(budget, total_expenses)

    display_summary(
        user,
        current_month,
        currency,
        budget,
        total_expenses,
        balance
    )


if __name__ == "__main__":
    main()
```
